```python
import subprocess
import matplotlib
import matplotlib.pyplot as plt
import re
#import os 

def ping_it ():
    
    rslt = subprocess.check_output("ping -n 100  www.google.com" ,shell=True) 
    print(  rslt)
    time = re.findall("time=[0-9][0-9][0-9]|time=[0-9][0-9]|time=[0-9]", str(rslt))
    print(time)
    lenth = len(rslt)
    mod =[]
    for x in time:
        mod.append(int(re.sub(r'time=','',x)))
    print(mod)
    
    plt.hist(mod, bins=[0,10 ,20 ,30,35,40,45, 50,80])
    
    plt.title('Google Ping Responses Time')
    plt.xlabel('no of pings')
    plt.ylabel('Pings reponse time')
    plt.grid(axis='y', alpha=0.75)
    plt.show()

     
ping_it ()
```
